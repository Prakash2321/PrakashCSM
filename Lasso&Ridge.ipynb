```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets,linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
diabetes=datasets.load_diabetes()
X=diabetes.data[:,np.newaxis,2]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,diabetes.target,test_size=0.2,random_state=42)
```

```python
regr=linear_model.LinearRegression()
regr.fit(X_train,y_train)
```

```python
y_pred=regr.predict(X_test)
y_test.sort()
y_pred.sort()
plt.plot(y_test,label="Actual Values",color='red')
plt.plot(y_pred,color='green',label="Predicted Values")
plt.plot('Actual vs. Predicted Values in Linear Regression')
plt.show()
```

```python
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
plt.scatter(X_test, y_test, color='blue', label='Actual')
plt.scatter(X_test, y_pred, color='red', label='Predicted')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
```
